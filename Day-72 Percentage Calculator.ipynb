### 3. Code Implementation
Here's the complete code for the percentage calculator application using tkinter:

```python
import tkinter as tk
from tkinter import messagebox

def calculate_percentage():
    try:
        percent = float(entry_percent_value.get())
        whole = float(entry_whole_value.get())
        result = (percent / 100) * whole
        label_result.config(text=f"Result: {result:.2f}")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

def calculate_percentage_of():
    try:
        value = float(entry_value.get())
        total = float(entry_total.get())
        if total == 0:
            messagebox.showerror("Error", "Total cannot be zero.")
            return
        result = (value / total) * 100
        label_result.config(text=f"Percentage: {result:.2f}%")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

def calculate_percentage_change():
    try:
        old_value = float(entry_old_value.get())
        new_value = float(entry_new_value.get())
        if old_value == 0:
            messagebox.showerror("Error", "Old value cannot be zero.")
            return
        change = ((new_value - old_value) / old_value) * 100
        label_result.config(text=f"Percentage Change: {change:.2f}%")
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Create the main window
root = tk.Tk()
root.title("Percentage Calculator")

# Calculate percentage (what is X% of Y)
tk.Label(root, text="What is X% of Y?").grid(row=0, column=0, padx=10, pady=5, sticky=tk.W)
tk.Label(root, text="Percentage (%):").grid(row=1, column=0, padx=10, pady=5, sticky=tk.W)
entry_percent_value = tk.Entry(root)
entry_percent_value.grid(row=1, column=1, padx=10, pady=5)

tk.Label(root, text="Whole:").grid(row=2, column=0, padx=10, pady=5, sticky=tk.W)
entry_whole_value = tk.Entry(root)
entry_whole_value.grid(row=2, column=1, padx=10, pady=5)

btn_percentage = tk.Button(root, text="Calculate", command=calculate_percentage)
btn_percentage.grid(row=3, columnspan=2, pady=10)

# Calculate what percentage of (X is what percent of Y?)
tk.Label(root, text="X is what percent of Y?").grid(row=4, column=0, padx=10, pady=5, sticky=tk.W)
tk.Label(root, text="Value:").grid(row=5, column=0, padx=10, pady=5, sticky=tk.W)
entry_value = tk.Entry(root)
entry_value.grid(row=5, column=1, padx=10, pady=5)

tk.Label(root, text="Total:").grid(row=6, column=0, padx=10, pady=5, sticky=tk.W)
entry_total = tk.Entry(root)
entry_total.grid(row=6, column=1, padx=10, pady=5)

btn_percentage_of = tk.Button(root, text="Calculate", command=calculate_percentage_of)
btn_percentage_of.grid(row=7, columnspan=2, pady=10)

# Calculate percentage increase/decrease
tk.Label(root, text="Percentage Increase/Decrease:").grid(row=8, column=0, padx=10, pady=5, sticky=tk.W)
tk.Label(root, text="Old Value:").grid(row=9, column=0, padx=10, pady=5, sticky=tk.W)
entry_old_value = tk.Entry(root)
entry_old_value.grid(row=9, column=1, padx=10, pady=5)

tk.Label(root, text="New Value:").grid(row=10, column=0, padx=10, pady=5, sticky=tk.W)
entry_new_value = tk.Entry(root)
entry_new_value.grid(row=10, column=1, padx=10, pady=5)

btn_percentage_change = tk.Button(root, text="Calculate", command=calculate_percentage_change)
btn_percentage_change.grid(row=11, columnspan=2, pady=10)

# Result display
label_result = tk.Label(root, text="Result will be displayed here.")
label_result.grid(row=12, columnspan=2, pady=10)

# Run the application
root.mainloop()
```
