## Data Visualization with Pandas
### Introduction:
Pandas integrates well with visualization libraries like Matplotlib and Seaborn to easily create plots.

### Step 1: Plotting Data with Pandas
You can create plots directly using pandasâ€™ built-in plotting features.

```python
import pandas as pd

# Reading a CSV file
df = pd.read_csv('samplefile.csv', delimiter= ';')

# Displaying the first few rows
print(df.head())
df
```

```python
import matplotlib.pyplot as plt

# Plotting a histogram of a column
df['Age'].plot(kind='hist')
plt.show()
```

### Visualizing all data

```python
df.plot()

plt.show()
```

### Step 2: Customizing Plots
You can customize plots using parameters in pandas or through Matplotlib.

```python
# Customizing a plot
df['Salary($)'].plot(kind='line', color='red', title='Salary over Time')
plt.show()
```

### Scatter Plot

```python
df.plot(kind = 'scatter', x = 'Age', y = 'Salary($)')

plt.show()
```

### Step 3: Advanced Visualizations
For more complex visualizations, you can use Seaborn.

```python
import seaborn as sns

# Creating a boxplot
sns.boxplot(x='Sex', y='Salary($)', data=df)
plt.show()
```
