### Step 1: Handling Missing Values (NaN)
Pandas uses NaN to represent missing values. You can detect missing values with isnull() and fill them using fillna().

```python
import pandas as pd

# Reading a CSV file
df = pd.read_csv('samplefile.csv', delimiter= ';')

# Displaying the first few rows
print(df.head())
df
```

```python
# Detecting missing values
missing_values = df.isnull()
missing_values
```

### Filling missing values with a specific value

```python
data_filled = df.fillna(0)
data_filled
```

### Specify Columns to fill

```python
age_filled = df['Age'].fillna(0)
age_filled
df
```

### Step 2: Dropping Missing Values
You can remove rows or columns that contain missing values with dropna().

```python
# Dropping rows with missing values
data_dropped = df.dropna()
data_dropped


```

```python
# Dropping columns with missing values
data_dropped_columns = df.dropna(axis=1)
data_dropped_columns 
```

### Step 3: Data Cleaning Techniques
Additional cleaning like renaming columns and changing data types can be done as follows:

```python
# Renaming columns
df.rename(columns={'Name': 'FirstName'}, inplace=True)
df

```
